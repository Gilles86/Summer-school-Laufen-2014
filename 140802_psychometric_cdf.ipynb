## Import relevant libraries

```python
import scipy as sp
from scipy import io
import pandas
from pprint import pprint
```

#Load data

```python
# Load matlab .mat-file using scipy into numpy-format
matlab = sp.io.loadmat('/Users/Gilles/Downloads/Ludwig2012.mat')

# Set up empty dataframe
data = pandas.DataFrame()

# Iterate over possible subjects
for pp in np.arange(1,9):

    # Get the first part of the data (condition = 'matched')
    data1 = pandas.DataFrame(matlab['P%d' % pp][0][0][0]) # Ugly...
    data1.columns = ['intensity', 'detected', 'n_trials'] # Column names
    data1['condition'] = 'matched'                         
    data1['subject_id'] = pp
    
    # Get second part of the data for this subject
    data2 = pandas.DataFrame(matlab['P%d' % pp][0][0][1])
    data2.columns = ['intensity', 'detected', 'n_trials']
    data2['condition'] = 'onset'
    data2['subject_id'] = pp
    
    # Concatenate the current entire dataset with the two sub-datasets
    data = pandas.concat((data, data1, data2))
    
# Create a new column with the 
data['prop'] = data.detected / data.n_trials
data = data.set_index(np.arange(data.shape[0]))
```

##Set up likelihood function

```python
def give_cdf_likelihood(data, pars):
    mu = pars[0]
    sigma = pars[1]
    ps = sp.stats.norm(mu, sigma).cdf(data.intensity)
    likelihood = sp.stats.binom(data.n_trials, ps).pmf(data.detected)
    return -np.sum(np.log(likelihood))
```

#Simplest model, same $\mu$ and $\sigma$

First we make a function that takes as input a dataframe, and fits the likelihood using fmin

```python
def fit_dataframe(d):
    # Do optimization using likelihood_cdf, start_params are [60, 15], return full output, 
    # including latest likelihood
    pars, ll, _, _, _ = sp.optimize.fmin(lambda pars: give_cdf_likelihood(d, pars), [60, 15], disp=0, full_output=True)
    
    # Return a new dataframe, including parameters and likelihood
    return pandas.DataFrame([list(pars) + [ll]], columns=['mu_simple','sigma_simple', 'll_simple'])
```

Now we split the data on the factor "subject_id" and apply the function we just wrote. This is super-easy: we don't need any nasty for-loops!

```python
d_simple = data.groupby(['subject_id']).apply(fit_dataframe)
```

That's it: now we have a dataframe with as rows the unique subject_id's, and as content the found parameters

```python
d_simple
```

Now we calculate BIC and AIC

```python
d_simple['bic_simple'] = 2 * d_simple['ll_simple'] + 2 * np.log(data.groupby('subject_id').n_trials.sum()) #n_trials
d_simple['aic_simple'] = 2 * d_simple['ll_simple'] + 2 * 2
d_simple
```

## Middle model: Different $\mu$, same $\sigma$

If we want to fit different $\mu$s but the sam $\sigma$, we can not split only on subject anymore, we have to write a slightly more complicated dataframe-fitting routine

```python
def fit_dataframe(d):
    pars, ll, _, _, _ = sp.optimize.fmin(lambda pars: 
                                         give_cdf_likelihood(d[d.condition == 'matched'], pars[[0,2]]) +
                                         give_cdf_likelihood(d[d.condition == 'onset'], pars[[1,2]]), 
                                                             [60, 60, 15], 
                                                             disp=0,
                                                             full_output=True)
    
    return pandas.DataFrame([list(pars) + [ll]], columns=['mu1_mixed','mu2_mixed', 'sigma_mixed', 'll_mixed'])

d_mixed = data.groupby(['subject_id']).apply(fit_dataframe)
```

## Most complex model: different $\mu$s, different $\sigma$s

```python
def fit_dataframe(d):
    pars, ll, _, _, _ = sp.optimize.fmin(lambda pars: 
                                         give_cdf_likelihood(d[d.condition == 'matched'], pars[[0,1]]) +
                                         give_cdf_likelihood(d[d.condition == 'onset'], pars[[2,3]]), 
                                                             [60, 15, 60, 15], 
                                                             disp=0,
                                                             full_output=True)
    
    return pandas.DataFrame([list(pars) + [ll]], columns=['mu1_complex', 'sigma1_complex', 'mu2_complex', 'sigma2_complex', 'll_complex'])

d_complex = data.groupby(['subject_id']).apply(fit_dataframe)
```

Write a function to get the minimum of a list of columns

```python
def get_min(d, columns):
    columns = np.array(columns)
    return columns[np.argmin(d[columns].values, 1)]
```

##Hierarchical Bayes

Now we're gonna use the Kabuki toolset of Thomas Wiecki 
(https://github.com/hddm-devs/kabuki) to set up a Hierarchical Bayesian Model, where we assume the parameters are drawn from hyperdistributions

```python
# import pymc
import pymc as pm
from collections import OrderedDict

# This is the likelihood function without the - before the log-likelihood
def give_cdf_likelihood(value, mu, sigma):
    ps = sp.stats.norm(mu, sigma).cdf(value[:, 0])
    likelihood = sp.stats.binom(value[:, 2], ps).pmf(value[:, 1])
    
    return np.sum(np.log(likelihood))


# We create a class to wrap the likelihood function
CDF_like = pm.stochastic_from_dist(name="CDF_like",
                                   logp=give_cdf_likelihood,
                                   dtype=np.float,
                                   mv=False)


# An another class to make it possible to include the likelihood into a
# Hierarchical model
class KnodeNormCDF(Knode):
    def create_node(self, node_name, kwargs, data):
#        print 'niger', kwargs
        kwargs['value'] = data[['intensity', 'detected', 'n_trials']]
        return self.pymc_node(name=node_name, **kwargs)

# Hierarhical Model-class
# Stuff below this is hacked together, adapting
# https://github.com/twiecki/stopsignal/blob/master/stopsignal/stopsignal.py
class PsychometricCdf(Hierarchical):

    def __init__(self, data, **kwargs):
        self.group_only_nodes = kwargs.pop('group_only_nodes', ())
        self.std_depends = kwargs.get('std_depends', False)

        super(PsychometricCdf, self).__init__(data, **kwargs)

    def create_ss_knode(self, knodes):
        ss_parents = OrderedDict()
        ss_parents['mu'] = knodes['mu_bottom']
        ss_parents['sigma'] = knodes['sigma_bottom']

        cdf_like = KnodeNormCDF(CDF_like, 'norm_cdf', col_name='prop', depends=['subj_idx'], observed=True, mu=ss_parents['mu'], sigma=ss_parents['sigma'])        

        return [cdf_like]

    def create_knodes(self):
        knodes = OrderedDict()

        knodes.update(self._create_family_trunc_normal('mu', lower=1e-3, upper=100, value=60, std_lower=0.01, std_upper=300, std_value=50))
        knodes.update(self._create_family_trunc_normal('sigma', lower=1e-3, upper=100, value=60, std_lower=0.01, std_upper=300, std_value=50))

        likelihoods = self.create_ss_knode(knodes)
        
        return knodes.values() + likelihoods

    
    def _create_family_trunc_normal(self, name, value=0, lower=None,
                                   upper=None, std_lower=1e-10,
                                   std_upper=100, std_value=.1):
        """Similar to _create_family_normal() but creates a Uniform
        group distribution and a truncated subject distribution.

        See _create_family_normal() help for more information.

        """
        knodes = OrderedDict()

        if self.is_group_model and name not in self.group_only_nodes:
            g = Knode(pm.Uniform, '%s' % name, lower=lower,
                      upper=upper, value=value, depends=self.depends[name])

            depends_std = self.depends[name] if self.std_depends else ()
            std = Knode(pm.Uniform, '%s_std' % name, lower=std_lower,
                        upper=std_upper, value=std_value, depends=depends_std)
            tau = Knode(pm.Deterministic, '%s_tau' % name,
                        doc='%s_tau' % name, eval=lambda x: x**-2, x=std,
                        plot=False, trace=False, hidden=True)
            subj = Knode(pm.TruncatedNormal, '%s_subj' % name, mu=g,
                         tau=tau, a=lower, b=upper, value=value,
                         depends=('subj_idx',), subj=True, plot=self.plot_subjs)

            knodes['%s'%name] = g
            knodes['%s_std'%name] = std
            knodes['%s_tau'%name] = tau
            knodes['%s_bottom'%name] = subj

        else:
            subj = Knode(pm.Uniform, name, lower=lower,
                         upper=upper, value=value,
                         depends=self.depends[name])
            knodes['%s_bottom'%name] = subj

        return knodes    
```
