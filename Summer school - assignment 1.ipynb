In this Python notebook we replicate the Monte Carlo simulations performed by Boneau (1960). In this paper, Boneau investigated the behavior of the t-test when certain assumptions of the t-test are violated, specifically different sample sizes, variances and distribution shapes.

```python
# Import stats packages
import scipy as sp
from scipy import stats
```

# Create parameters

First we set up the different parameters. There is one tricky case: when both the number of samples _and_ the variances differ, the order of things matter.

```python
pars1 = 
        ({'n1':5, 'n2':5,'V1':1,'V2':1},    # Equal variances 1
         {'n1':5, 'n2':15,'V1':1,'V2':1},
         {'n1':15, 'n2':15,'V1':1,'V2':1},

         {'n1':5, 'n2':5,'V1':1,'V2':4},   # Differnt variances (1 and 4)
         {'n1':5, 'n2':15,'V1':1,'V2':4},  
         {'n1':5, 'n2':15,'V1':4,'V2':1},  # Special case#
         {'n1':15, 'n2':15,'V1':1,'V2':4},

         {'n1':5, 'n2':5,'V1':4,'V2':4},   #Equal variances 4
         {'n1':5, 'n2':15,'V1':4,'V2':4},
         {'n1':15, 'n2':15,'V1':4,'V2':4})

    
# Now we have to do all these combinations for the three different distribution families. 
# For this we use a product iteration-function.

# Set up list
pars = []

# Generate all combinations
for par, distr in itertools.product(pars1, ['normal', 'expon', 'uniform']):
    par.update({'distribution':distr})
    pars.append(par.copy())

    
# Throw everything in a dataframe
import pandas
pars = pandas.DataFrame(pars)
```

Now we're gonna do bla

We set up af function that can simulate a _single_ experiment: we just grab two datasets with given $n$ and $V$

```python
def experiment(stds, ns, distr):
    
    # Which distribution?
    if distr == 'normal':
        data1 = sp.stats.norm(loc=0, scale=stds[0]).rvs(ns[0])   # Dataset 1
        data2 = sp.stats.norm(loc=0, scale=stds[1]).rvs(ns[1])   # Dataset 2
        
    elif distr == 'expon':
        data1 = sp.stats.expon(loc=0, scale=stds[0]).rvs(ns[0])
        data2 = sp.stats.expon(loc=0, scale=stds[1]).rvs(ns[1])
    
    elif distr == 'uniform':
        data1 = sp.stats.uniform(loc=0, scale=stds[0]).rvs(ns[0])
        data2 = sp.stats.uniform(loc=0, scale=stds[1]).rvs(ns[1])
        
    return sp.stats.ttest_ind(data1, data2)
    
```

Now we set up a second function that can repeat an experiment $n$ times, given the parameters _kwargs_

```python
def do_experiments(n=1000, **kwargs):
    ts = []
    ps = []

    for i in np.arange(1000):
        t, p = experiment(**kwargs)
        ts.append(t)
        ps.append(p)

    ts = np.array(ts)
    ps = np.array(ps)
    
    return ts, ps
```

Let's apply it to all the parameter sets and store the results

```python
results = pandas.DataFrame()

for i, row in pars.iterrows():
    print '%s/%s' % (i, len(pars))
    t, p = do_experiments(100, stds=[np.sqrt(row['V1']),np.sqrt(row['V2'])], ns=[row['n1'], row['n2']], distr=row.distribution)
    
    df = pandas.DataFrame(np.array([t, p]).T, columns=['t', 'p'])
    
    for key in row.keys():
        df[key] = row[key]
    
    results = pandas.concat((results, df))
    
```

Now we can have a look at the empirical proportion of Type II errors. Remember that these are supposed to be 0.05

```python
results.groupby(['distribution', 'n1', 'n2', 'V1', 'V2']).apply(lambda d: np.mean(d.p < 0.05))
```

# Conclusions

The t-test is surprisingly robust against some violations of its assumptions, especially when both sample sets are normally distributed.

_However_, things completely break down when the data is uniformly distributed and the variance of the two datasets is different.
