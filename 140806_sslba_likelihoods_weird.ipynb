Wrapper function to create two-choice stop data in pandas DataFrame

```python
def create_data(V=2,
                v=1,
                ter=0.1,
                A=0.3,
                b=1,
                Vs=5,
                n=500):
    
    
    class testModel1_paramspec(pr.Parameters):
        parnames=["A","V","ter","v", 'Vs']
        lower   =[0,-inf,0,-inf, -inf]
        upper   =[inf,inf,inf,inf, inf]
        modelstr=""

    def __reduce__(self):
        return (_load_classinst_from_string, (self.modelstr, self.__class__.__name__, self.__dict__))

    class testModel1(pr.pSSLBA):

        paramspec=testModel1_paramspec
        modelstr=""

        def __init__(self, pars=None):
            self.design=pr.Design([{'stimulus': ['left', 'right']}], ['left', 'right'], 'stimulus', name='singlego')
            self.set_mixing_probabilities(0,0)

            if pars!=None:
                if not isinstance(pars, self.paramspec):
                    pars=self.paramspec(pars)
                self.set_params(pars)
            else:
                self.set_params(self.__class__.paramspec().random())

        def copy(self):
            m=self.__class__(self.params)
            m.modelstr=self.modelstr
            return m

        def __reduce__(self):
            return (_load_classinst_from_string, (self.modelstr, self.name(), self.__dict__))

        def set_params(self, pars):
            self.params=pars
            _go_acc=[ [None for resp in range(self.design.nresponses())] for cond in range(self.design.nconditions())]
            _stop_acc=[ None for cond in range(self.design.nconditions())]

            A=pars.A
            V=pars.V
            ter=pars.ter
            v=pars.v
            Vs=pars.Vs

            _go_acc[0][0]=self.accumulator_type(A=A,ter=ter,b=b,sv=1,v=V, name='go-'+':'.join(self.design.condidx(0)))
            _go_acc[0][1]=self.accumulator_type(A=A,ter=ter,b=b,sv=1,v=v, name='go-'+':'.join(self.design.condidx(0)))
            _go_acc[1][0]=self.accumulator_type(A=A,ter=ter,b=b,sv=1,v=v, name='go-'+':'.join(self.design.condidx(1)))
            _go_acc[1][1]=self.accumulator_type(A=A,ter=ter,b=b,sv=1,v=V, name='go-'+':'.join(self.design.condidx(1)))


            _stop_acc[0]=self.accumulator_type(A=A,ter=ter,b=b,sv=1,v=Vs, name='stop-'+':'.join(self.design.condidx(0)))
            _stop_acc[1]=self.accumulator_type(A=A,ter=ter,b=b,sv=1,v=Vs, name='stop-'+':'.join(self.design.condidx(1)))


            self.set_accumulators(_go_acc, _stop_acc)

            ## mixing probabilities
            _pgf=[0 for cond in range(self.design.nconditions())]
            _ptf=[0 for cond in range(self.design.nconditions())]

            _pgf[0]=0
            _pgf[1]=0


            _ptf[0]=0
            _ptf[1]=0


            self.set_mixing_probabilities(_pgf, _ptf)
            
    mod =  testModel1([A, V, ter, v, Vs])
    #mod =  testModel1({'A':1, 'V':3, 'ter':0.4, 'v':1, 'Vs':5})

    #display(mod.params)
    dsim=mod.simulate(n) # 500 trials per condition

    
    data = dsim.as_dataframe()
    data.head()

    data['stim'] = data.condition.map({'left':0, 'right':1})


    dsim.check()
    
    return data, mod
```

SSLBA likeilhood wrapper

```python
def sslba_likelihood(value=None,
                     V=2.,
                    v=1.,
                    ter=.1,
                    A=0.3,
                    b=1.,
                    Vs=5.,
                    sv=1.,
                    stop_A=None,
                    stop_b=None,
                    stop_sv=None,
                    stop_ter=None,
                    pgf=0.,
                    ptf=0.,
                    ):
    
    
    if stop_sv == None:
        stop_sv = sv
    
    if stop_ter == None:
        stop_ter = ter
        
    if stop_A == None:
        stop_A = A
        
    if stop_b == None:
        stop_b = b
                  
    
    
    
    n_trials = value.shape[0]
    L = np.zeros(n_trials)
    
    #print value
    
    crace.sslba_loglikelihood(2, # n_conditions
                          2, # n_responses
                          n_trials, # n_trials
                          np.array(value[:, 2], dtype=int32), # condition
                          np.array(value[:, 1], dtype=int32), # responses
                          np.array(value[:, 3], dtype=float), # RT
                          np.array(value[:, 4], dtype=float), # SSD
                          np.array([V, v, v, V]), # go_v
                          np.array([ter, ter] * 2), # go_ter
                          np.array([A, A] * 2), # go_A
                          np.array([b, b] * 2), # go_b
                          np.array([sv, sv] * 2), # go_sv                              
                          np.array([Vs] * 2), 
                          np.array([stop_ter] * 2),
                          np.array([stop_A] * 2),
                          np.array([stop_b] * 2),
                          np.array([stop_sv] * 2),
                          np.array([pgf] * 2),
                          np.array([ptf] * 2),
                          L)
    
    L = np.sum(L)

    
    return L
```
